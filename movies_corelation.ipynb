```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep='\t', names = column_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv('Movie_Id_Titles')
movie_titles.head()
```

```python
df = pd.merge(df, movie_titles, on='item_id')
df.head()
```

```python
sns.set_style('white')
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
plt.figure(figsize=(10,6))
ratings['num of ratings'].hist(bins=70)
```

```python
plt.figure(figsize=(10,6))
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating', y='num of ratings', data = ratings, alpha = 0.5)
```

```python
moviemat = df.pivot_table(index='user_id', columns='title', values = 'rating')
moviemat.head()
```

```python
ratings.sort_values('num of ratings', ascending=False).head(10)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
similiar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similiar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similiar_to_starwars, columns=['Corelation'])
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('Corelation', ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
corr_starwars.head(10)
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Corelation', ascending=False).head(10)
```

```python
corr_liarliar = pd.DataFrame(similiar_to_liarliar, columns=['Corelation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('Corelation', ascending=False).head(10)
```
